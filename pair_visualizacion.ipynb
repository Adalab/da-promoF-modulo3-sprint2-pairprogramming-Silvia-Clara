1 - ¿Cómo se distribuye la tasa de natalidad en el conjunto de datos? ¿Hay algún patrón evidente?

```python
#Histogarma

sns.histplot(x = "birthrate", data= df_graficas, bins= 15)

plt.xlabel("Tasa Natalidad")

plt.ylabel("Conteo por países")

plt.title("Distribución de la natalidad", fontsize = 20)

plt.gca().spines['right'].set_visible(False)
plt.gca().spines["top"].set_visible(False)

;
```

En términos generales cuánto más alta es la tasa de nataliad el número de paises que poseen este valor es menor. 

2 - ¿Existe una relación aparente entre la tasa de mortalidad infantil y la esperanza de vida? Proporciona una breve descripción de los patrones que observes.

```python
sns.regplot(x = "infantmortality", 
            y = "lifeexpectancy", 
            data = df_graficas, 
            marker = ".", 
            line_kws = {"color": "teal", "linewidth": 2}, # cambiamos el color y el grosor de la linea de tendencia
            scatter_kws = {"color": "mediumorchid", "s": 5} # cambiamos el color y el tamaño de los puntosdel scaterplot
            )

plt.xlabel("Mortalidad infantil")

plt.ylabel("Esperanza de vida")

plt.title("Relación entre mortalidad infantil y esperanza de vida", fontsize = 20)

plt.gca().spines['right'].set_visible(False)
plt.gca().spines["top"].set_visible(False)
;
```

Cuanto menor es la mortalidad infantil mayor es la esperanza de vida.  Son inversamente proporcionales. 

3 - ¿Cuál es el continente con la mayor cantidad de países en el conjunto de datos? Proporciona el porcentaje de países en ese continente.

```python
df_graficas.columns
```

```python
pais_continente = df_graficas.groupby("Continent")["abbreviation"].count().reset_index()
pais_continente
```

```python
plt.pie("abbreviation", labels= "Continent",
        data = pais_continente, 
        autopct=  '%1.1f%%', 
        colors = ["c", "cadetblue", "mediumseagreen", "seagreen", "teal", "turquoise", "lightseagreen"],
        textprops={'fontsize': 8}, 
        startangle=90)


plt.title("Porcentaje de países por continente", fontsize = 20)

;
```

5 - ¿Existen relaciones visibles entre la tasa de natalidad y las siguientes variables: tasa de mortalidad infantil, esperanza de vida, densidad de población y población urbana? Usa un subplot. Describe cualquier patrón que encuentres.

```python
df_graficas["infantmortality"].info()
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (20, 5))

sns.regplot(x = "birthrate", 
            y = "infantmortality", 
            data = df_graficas, 
            marker = ".", 
            line_kws = {"color": "black", "linewidth": 1}, 
            scatter_kws = {"color": "teal", "s": 1}, 
            ax= axes[0]
            )

axes[0].set_xlabel("Mortalidad infantil")

axes[0].set_ylabel("Esperanza de vida")

axes[0].set_title("Relación entre mortalidad infantil y esperanza de vida", fontsize = 20)

axes[0].spines['right'].set_visible(False)
axes[0].spines["top"].set_visible(False)

;
```
