1 - ¿Cómo se distribuye la tasa de natalidad en el conjunto de datos? ¿Hay algún patrón evidente?

```python
#Histogarma

sns.histplot(x = "birthrate_k", data= df_graficas, bins= 15)

plt.xlabel("Tasa Natalidad")

plt.ylabel("Conteo por países")

plt.title("Distribución de la natalidad", fontsize = 20)

plt.gca().spines['right'].set_visible(False)
plt.gca().spines["top"].set_visible(False)

;
```

2 - ¿Existe una relación aparente entre la tasa de mortalidad infantil y la esperanza de vida? Proporciona una breve descripción de los patrones que observes.

```python
sns.regplot(x = "infantmortality_k", 
            y = "lifeexpectancy_i", 
            data = df_graficas, 
            marker = ".", 
            line_kws = {"color": "teal", "linewidth": 2}, # cambiamos el color y el grosor de la linea de tendencia
            scatter_kws = {"color": "mediumorchid", "s": 5} # cambiamos el color y el tamaño de los puntosdel scaterplot
            )

plt.xlabel("Mortalidad infantil")

plt.ylabel("Esperanza de vida")

plt.title("Relación entre mortalidad infantil y esperanza de vida", fontsize = 20)

plt.gca().spines['right'].set_visible(False)
plt.gca().spines["top"].set_visible(False)
;
```

Cuanto menor es la mortalidad infantil mayor es la esperanza de vida.  Son inversamente proporcionales.

6 - ¿Cuáles son los 10 países con la mayor área forestal?

```python
#ordenamos el DF de mayor a menor los paises con mayor area forestal
df_sorted = df_graficas.sort_values(by='forestedarea', ascending=False)

df_10_paises_a_forestal= df_sorted.head(10)
df_10_paises_a_forestal
```

```python
# Crear el gráfico de barras con Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(x='forestedarea', y='country', data=df_10_paises_a_forestal, palette='viridis', hue='country', legend = False)
plt.xlabel('Área Forestal', fontsize=12)
plt.ylabel('Países', fontsize=12 )
plt.title('Top 10 Países con Mayor Área Forestal',fontsize=16 )
;

```

8 - ¿Cómo se relacionan el CPI y el precio de la gasolina? Crea un scatter plot que muestre esta relación y agrega una línea para visualizar la tendencia. Ajusta los valores del CPI entre 100 y 300.

```python
# Ajustamos los valores del CPI entre 100 y 300.
cpi_100_300 = df_graficas[(df_graficas['cpi_k'] >= 100) & (df_graficas['cpi_k'] <= 300)]

# Crear el scatter plot con línea de tendencia
plt.figure(figsize=(12, 8))
sns.regplot(x='cpi_k', y='gasolineprice_i', data = cpi_100_300, scatter_kws={'s': 2}, line_kws={'color': 'red'})
plt.xlabel('Índice de Precios al Consumidor (CcpiPI)', size= 12)
plt.ylabel('Precio de la Gasolina', size= 12)
plt.title('Relación entre CPI y Precio de la Gasolina', size= 16)
plt.grid(True)
plt.show()
```

No hay relación aparente entre el precio de gasolina y el índice de precios al consumidor.
